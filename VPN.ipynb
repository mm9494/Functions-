```python
df = df[df['GDP '] != '#DIV/0!'] # Remove #DIV/0! values
df = df[(df['GDP '] != 0) & (df['FOTN'] != 0)] # Remove 0 values
df['Adoption'] = df['Adoption'].apply(lambda x: x.replace('%', '')) # Remove '%' from Adoption string values
df
```

```python
df = df.apply(pd.to_numeric, errors='raise')
df['GDP_normalized'] = (df['GDP '] - df['GDP '].mean()) / df['GDP '].std()
df['FOTN_normalized'] = (df['FOTN'] - df['FOTN'].mean()) / df['FOTN'].std()
X = df[['GDP_normalized','FOTN_normalized',]]  # Replace with your independent variables
y = df['Adoption']  # Replace with your dependent variable


# X = sm.add_constant(X)
# model = sm.OLS(y, X).fit()
# print(model.summary())

# df[df['GDP '] != '#DIV/0!']
```

```python
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```
